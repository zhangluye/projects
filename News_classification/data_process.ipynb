```python
import numpy as np
import re
import pandas as pd

trainset_path = './data/trainset.xlsx'
df_caijing = pd.read_excel(trainset_path,sheet_name='财经')  #8597
df_fangchan = pd.read_excel(trainset_path,sheet_name='房产') #200
df_jiaoyu = pd.read_excel(trainset_path,sheet_name='教育') #500
df_keji = pd.read_excel(trainset_path,sheet_name='科技') #830
df_junshi = pd.read_excel(trainset_path,sheet_name='军事') #158
df_qiche = pd.read_excel(trainset_path,sheet_name='汽车') #647
df_tiyu = pd.read_excel(trainset_path,sheet_name='体育') #1200
df_youxi = pd.read_excel(trainset_path,sheet_name='游戏') #1300
df_yule = pd.read_excel(trainset_path,sheet_name='娱乐') #1200
df_qita = pd.read_excel(trainset_path,sheet_name='其他')

print(' ===============================  data loading success! ===================================')
```

```python
import jieba
stopwords_path = './data/stop_words.txt'
stopwords = [line.strip() for line in open(stopwords_path,encoding='UTF-8').readlines()]

#处理标点、数字
def punct_handling(inputstr):
    remove_str = "[\s+\.:\!-\/_,$%\[\]^\)*(+\"\']+|[+——！1234567890·，。？?、~@#￥%……&*（）` | \" ~〜 “ １　２　３　４　５　６　７　８　９　０”【】 ※]+"
    punc = "！？｡＂＃＄％＆＇（）＊＋，－／：；＜＝＞＠［＼］＾＿｀｛｜｝～｟｠｢｣､、〃》「」『』【】〔〕〖〗〘〙〚〛〜〝〞〟〰〾〿–—‘’‛“”„‟…‧﹏."
    punc = punc.encode('utf-8').decode("utf-8")
    string1 = re.sub(remove_str,"",inputstr)
    string1 = re.sub(punc,"",string1)
    return string1


# 对句子进行中文分词
def seg_depart(sentence):
    # 对文档中的每一行进行中文分词
    sentence = punct_handling(sentence)
    sentence_depart = jieba.cut(sentence.strip())
    # 输出结果为outstr
    outstr = ''
    # 去停用词
    for word in sentence_depart:
        if word not in stopwords:
            if word != '\t':
                outstr += word
                outstr += " "
    return outstr

'''
    这里只选择了内容进行分词
'''


#存储分词结果
df_caijing['content_processed'] = df_caijing['content'].astype('str').apply(seg_depart)
df_fangchan['content_processed'] = df_fangchan['content'].astype('str').apply(seg_depart)
df_jiaoyu['content_processed'] = df_jiaoyu['content'].astype('str').apply(seg_depart)
df_keji['content_processed'] = df_keji['content'].astype('str').apply(seg_depart)
df_junshi['content_processed'] = df_junshi['content'].astype('str').apply(seg_depart)
df_qiche['content_processed'] = df_qiche['content'].astype('str').apply(seg_depart)
df_tiyu['content_processed'] = df_tiyu['content'].astype('str').apply(seg_depart)
df_yule['content_processed'] = df_yule['content'].astype('str').apply(seg_depart)
df_youxi['content_processed'] = df_youxi['content'].astype('str').apply(seg_depart)
#df_qita['content_processed'] = df_qita['content'].astype('str').apply(seg_depart)

print(' ===============================  participate complete! ===================================')
```

```python
import gensim
from gensim.models import Word2Vec
from gensim.models.word2vec import LineSentence

save_path = './data_processed/'
#获取词库
def get_words_texts(series,classes):
    x_texts = ''
    for i in range(series.shape[0]):
        x_texts = x_texts + series[i].strip() + ' '

    return x_texts.strip()

'''
    word2vec模型的参数：
    
    # Word2Vec第一个参数代表要训练的语料
    # sg=1 表示使用Skip-Gram模型进行训练
    # size 表示特征向量的维度，默认为100。大的size需要更多的训练数据,但是效果会更好. 推荐值为几十到几百。
    # window 表示当前词与预测词在一个句子中的最大距离是多少
    # min_count 可以对字典做截断. 词频少于min_count次数的单词会被丢弃掉, 默认值为5
    # workers 表示训练的并行数
    #sample: 高频词汇的随机降采样的配置阈值，默认为1e-3，范围是(0,1e-5)
'''

#生成word2vec模型
def get_w2vmodel(x_texts):
    filename = './data_processed/words.txt'
    textfile = open(filename, 'w')
    textfile.write(x_texts)
    textfile.close()
    
    words = open(filename, 'rb')
    w2vmodel = gensim.models.Word2Vec(LineSentence(words), sg=1, window=10, min_count=1, workers=15,sample=1e-3)
    model_path = './model/content_w2v.bin'
    w2vmodel.wv.save_word2vec_format(model_path, binary=True) 

#输入数据生成word2vec模型
def get_all_w2vmodel():
    fn_caijing = get_words_texts(df_caijing['content_processed'],'caijing')
    fn_fangchan = get_words_texts(df_fangchan['content_processed'],'fangchan')
    fn_jiaoyu = get_words_texts(df_jiaoyu['content_processed'],'jiaoyu')
    fn_keji = get_words_texts(df_keji['content_processed'],'keji')
    fn_junshi = get_words_texts(df_junshi['content_processed'],'junshi')
    fn_qiche = get_words_texts(df_qiche['content_processed'],'qiche')
    fn_tiyu = get_words_texts(df_tiyu['content_processed'],'tiyu')
    fn_youxi = get_words_texts(df_youxi['content_processed'],'youxi')
    fn_yule = get_words_texts(df_yule['content_processed'],'yule')
    #其他类 fn_qita
    x_texts = fn_caijing+" "+fn_fangchan+" "+fn_jiaoyu+" "+fn_keji+" "+fn_junshi+" "+fn_qiche+" "+fn_tiyu+" "+fn_youxi+" "+fn_yule
    get_w2vmodel(x_texts)

    #其他类 get_w2vmodel(fn_)
    
    
```

```python
get_all_w2vmodel()
print(' ===============================  training model completed! ===================================')
```

```python
news_num = 158 #每种新闻取多少条

def get_class_df(class_name,news_num):
    src_path = save_path + class_name + '_m1.csv'
    df = pd.read_csv(src_path, header=None, engine="python", encoding="utf-8")
    df = df[1:]#去掉第一行，因为是原列名
    df.columns = ['index','content','type','title','content_processed']
    df['type'] = class_name  #取一个新列
    new_df = df[["content_processed","type"]]
    return new_df[:news_num]

#生成训练数据集
def get_traindf(news_num):
    df_all = get_class_df('caijing',news_num)
    
    new_df = get_class_df('fangchan',news_num)
    df_all = pd.concat([df_all,new_df])
    new_df = get_class_df('jiaoyu',news_num)
    df_all = pd.concat([df_all,new_df])
    new_df = get_class_df('keji',news_num)
    df_all = pd.concat([df_all,new_df])
    new_df = get_class_df('junshi',news_num)
    df_all = pd.concat([df_all,new_df])
    new_df = get_class_df('qiche',news_num)
    df_all = pd.concat([df_all,new_df])
    new_df = get_class_df('tiyu',news_num)
    df_all = pd.concat([df_all,new_df])
    new_df = get_class_df('youxi',news_num)
    df_all = pd.concat([df_all,new_df])
    new_df = get_class_df('yule',news_num)
    df_all = pd.concat([df_all,new_df])
    
    df_labels = pd.get_dummies(df_all['type'])  #生成one_hot
    df_res = pd.concat([df_all,df_labels],axis = 1)
    
    dir_path = save_path + 'train_data.csv'
    df_res.to_csv(dir_path)
    return df_res
    
```
